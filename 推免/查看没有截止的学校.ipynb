```python
filname = '双非'
with open('{0}.md'.format(filname),'r',encoding='utf-8') as fp:
    text = fp.read()
text = text.split('\n')
while '' in text:
    text.remove('')



text_write = []

def compare_time(time1, time2):
    d1 = datetime.datetime.strptime(time1, '%Y-%m-%d')
    d2 = datetime.datetime.strptime(time2, '%Y-%m-%d')
    delta = d1 - d2
    #print("days is ", delta.days)
    # 没戏了
    if delta.days >= 1:
        return -1
    # 刚好今天
    elif delta.days == 0:
        return 0
    # 还有机会
    else:
        return 1

for each_line in text:
    import re
    pattern = re.compile('\d{4}-\d{1,2}-\d{1,2}')
    
    # 如果找到了日期
    if pattern.search(each_line) != None:
        ddl = pattern.search(each_line)[0]
        # 如果有很多天
        if compare_time(time_now, ddl) == -1:
            each_line = '~~' + each_line + '~~' 
            print(each_line)
        elif compare_time(time_now, ddl) ==0:
            each_line = '**' + each_line + '**' 
            print(each_line)
        elif compare_time(time_now, ddl) == 1:
            print(each_line)
            
    text_write.append(each_line)
        
with open('{0}_筛选.md'.format(filname),'w',encoding='utf-8') as fp:
    fp.write('\n'.join(text_write))
```
